Load & Explore the Data

```python
import pandas as pd

matches = pd.read_csv("C:/Users/DELL/Desktop/Data Ipl Project/archive (1)/matches.csv")
deliveries = pd.read_csv("C:/Users/DELL/Desktop/Data Ipl Project/archive (1)/deliveries.csv")

# Preview data
matches.head()
deliveries.head()
```

```python
matches.info()
deliveries.info()
```

```python
matches['date'] = pd.to_datetime(matches['date'])
```

```python
matches['date'].head()

```

FIX & CLEAN season COLUMN

```python
matches['season'] = matches['season'].astype(str)
matches['season'] = matches['season'].apply(lambda x: x[:4])  # Keep only first 4 characters
matches['season'] = matches['season'].astype(int)
```

CLEAN DATA

```python
# Check missing values
matches.isnull().sum()
deliveries.isnull().sum()

# Convert 'date' to datetime
matches['date'] = pd.to_datetime(matches['date'])

# Remove no-result or abandoned matches
matches = matches[matches['result'] != 'no result']

matches
```

 FEATURE ENGINEERING

```python
# Capitalize the result column directly
matches['win_type'] = matches['result'].str.capitalize()
matches
```

```python
matches['toss_match_win'] = matches['toss_winner'] == matches['winner']
matches
```

```python
deliveries = deliveries.merge(matches[['id', 'season']], left_on='match_id', right_on='id')
deliveries
```

SEASONAL TRENDS & TEAM STATS

```python
matches['season'].value_counts().sort_index().plot(kind='bar', title='Matches per Season')
```

```python
matches['winner'].value_counts().plot(kind='bar', title='Overall Team Wins')
```

```python
toss_impact = matches['toss_match_win'].mean() * 100
print(f"Toss Winner wins the match {toss_impact:.2f}% of the time.")
```

PLAYER PERFORMANCE

```python
print(deliveries.columns.tolist())
```

```python
top_batsmen = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_batsmen.plot(kind='bar', title='Top 10 Batsmen (by Runs)', ylabel='Total Runs', xlabel='Batter', figsize=(10,5))
```

```python
wickets = deliveries[deliveries['dismissal_kind'].notnull()]
top_bowlers = wickets.groupby('bowler')['player_dismissed'].count().sort_values(ascending=False).head(10)
top_bowlers.plot(kind='bar', title='Top 10 Bowlers')
```

```python
balls_faced = deliveries.groupby('batter').count()['ball']
total_runs = deliveries.groupby('batter')['batsman_runs'].sum()
strike_rate = (total_runs / balls_faced) * 100
strike_rate = strike_rate[balls_faced > 200].sort_values(ascending=False).head(10)
strike_rate.plot(kind='bar', title='Top 10 Strike Rates (min 200 balls)')
```

VENUE ANALYSIS

```python
matches['venue'].value_counts().head(10).plot(kind='bar', title='Top Venues by Match Count')
```

```python
venue_scores = deliveries.groupby(['match_id', 'inning'])['total_runs'].sum().reset_index()
venue_scores = venue_scores.merge(matches[['id', 'venue']], left_on='match_id', right_on='id')
venue_avg = venue_scores.groupby('venue')['total_runs'].mean().sort_values(ascending=False).head(10)
venue_avg.plot(kind='barh', title='Avg Innings Total by Venue')
```

SCORING PATTERN OVER SEASONS

```python
season_scores = deliveries.groupby(['match_id'])['total_runs'].sum().reset_index()
season_scores = season_scores.merge(matches[['id', 'season']], left_on='match_id', right_on='id')
avg_score_season = season_scores.groupby('season')['total_runs'].mean()

avg_score_season.plot(kind='line', marker='o', title='Avg Match Score per Season')
```
